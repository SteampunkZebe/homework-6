```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
import pandas as pd
import os
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import fetch_mldata
from sklearn import preprocessing
from sklearn.feature_selection import SelectPercentile
from sklearn.linear_model import LinearRegression
import math
#loading the data in and looking it over
los=pd.read_csv(os.path.join("LasVegasTripAdvisorReviews-Dataset (1).csv"))
print(los)
```

```python
#visualizing the data
y=los["Hotel stars"]
plt.plot(y)
```

```python
#train test split
X_train,X_test,y_train,y_test=train_test_split(los.iloc[:,2:4], y, test_size=0.25, random_state=42)
#select the model
from sklearn.ensemble import GradientBoostingClassifier
#tune hyperparameters
gbrt = GradientBoostingClassifier(random_state=42,n_estimators=10,max_depth=10,learning_rate=1)
gbrt.fit(X_train,y_train)
#test
print(gbrt.score(X_train,y_train))
print(gbrt.score(X_test,y_test))
```
