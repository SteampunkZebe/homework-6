```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
import pandas as pd
import os
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import fetch_mldata
from sklearn import preprocessing
from sklearn.feature_selection import SelectPercentile
from sklearn.linear_model import LinearRegression
import math

energy=pd.read_excel(os.path.join("Copy of ENB2012_data.xlsx"))
```

```python
#visualising the data 
plt.plot(energy["Y1"])
plt.plot(energy["Y2"])
```

```python
#creating a custom label thing for the tests 
label=[]
for x,y in zip(energy["Y1"],energy["Y2"]):
    if (x+y) < 22.31:
        label.append("low")
    else:
        label.append("high")
        
X_train,X_test,y_train,y_test=train_test_split(energy.iloc[:,:-3],label, test_size=0.25, random_state=42)
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(solver='lbfgs', random_state=0,hidden_layer_sizes=[10,10,10,10,10]).fit(X_train, y_train)
mlp.fit(X_train,y_train)
print("Accuracy on training set: {:.2f}".format(mlp.score(X_train, y_train)))
print("Accuracy on test set: {:.2f}".format(mlp.score(X_test, y_test)))
```
